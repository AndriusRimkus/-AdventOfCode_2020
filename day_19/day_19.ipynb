```python
import re

input_file = open("input.txt")
input = input_file.read().strip().split('\n')
input_file.close()

rules = {}
test_lines = []
start_collecting_test_lines = False;

for line in input:
    if line == '':
        start_collecting_test_lines = True
        continue
    
    if start_collecting_test_lines:
        test_lines.append(line)
        continue
    
    (ruleIndex, rule) = re.findall('(\d+): (.*)', line)[0]
    
    if '"' in rule:
        processed_rule = rule.replace('"', '')
    else:
        rule_parts = rule.split('|')
        processed_rule = []
        
        for rule_part in rule_parts:
            processed_rule.append([int(x) for x in re.findall('(\d+)', rule_part)])
    
    rules[int(ruleIndex)] = processed_rule
    
def build_regex(rule_index, parents = []):
    rule = rules[rule_index]
    
    if isinstance(rule[0], str):
        return rule[0]
    
    if (len(rule) == 1):
        result = ''
        
        for rule_instruction in rule[0]:
            if parents.count(rule_instruction) >= 5:
                return ''
            
            result += build_regex(rule_instruction, parents + [rule_index])
        
        return result
    
    if (len(rule) == 2):
        result = '('
        
        for rule_instruction in rule[0]:
            if parents.count(rule_instruction) >= 5:
                return ''
            
            result += build_regex(rule_instruction, parents + [rule_index])
            
        result += '|'
        
        for rule_instruction in rule[1]:
            if parents.count(rule_instruction) >= 5:
                return ''
            
            result += build_regex(rule_instruction, parents + [rule_index])
            
        result += ')'
        
        return result;
    
rule_regex = build_regex(0)

pattern = re.compile('^' + rule_regex + '$')

valid_count = 0

for test_line in test_lines:
    if pattern.match(test_line):
        valid_count += 1
        
print(valid_count)
```
