```python
import re
import math
import numpy as np
import itertools
import collections

input_file = open("input.txt")
input = input_file.read().strip().split('\n')
input_file.close()

player_1_cards = None
player_2_cards = None

def fill_player_decks():
    global player_1_cards
    global player_2_cards
    
    start_filling_player_2_cards = False
    
    player_1_cards = collections.deque([])
    player_2_cards = collections.deque([])
    
    for line in input:
        if len(line) == 0:
            continue

        if 'Player 2' in line:
            start_filling_player_2_cards = True

        if 'Player' in line:
            continue

        if start_filling_player_2_cards:
            player_2_cards.append(int(line))
        else:
            player_1_cards.append(int(line))

def determine_deck_score(deck):
    multi = 1
    score = 0

    while len(deck) > 0:
        score += deck.pop() * multi
        multi += 1
    
    return score
            
fill_player_decks()

# PART 1

while len(player_1_cards) > 0 and len(player_2_cards) > 0:
    player_1_card = player_1_cards.popleft()
    player_2_card = player_2_cards.popleft()
    
    if player_1_card > player_2_card:
        player_1_cards.extend([player_1_card, player_2_card])
    else:
        player_2_cards.extend([player_2_card, player_1_card])

if len(player_1_cards) == 0:
    print(determine_deck_score(player_2_cards))
else:
    print(determine_deck_score(player_1_cards))

# PART 2
    
def determine_winner(player_1_cards, player_2_cards):
    encountered_game_states = {}
    
    while len(player_1_cards) > 0 and len(player_2_cards) > 0:
        game_state = ''.join(map(str, player_1_cards)) + '|' + ''.join(map(str, player_2_cards))
        
        if game_state in encountered_game_states:
            return 'player 1'
        
        encountered_game_states[game_state] = True
        
        player_1_card = player_1_cards.popleft()
        player_2_card = player_2_cards.popleft()

        should_trigger_sub_game = len(player_1_cards) >= player_1_card and len(player_2_cards) >= player_2_card

        if should_trigger_sub_game:
            player_1_sub_game_cards = collections.deque(list(player_1_cards)[0:player_1_card])
            player_2_sub_game_cards = collections.deque(list(player_2_cards)[0:player_2_card])
            
            winner = determine_winner(player_1_sub_game_cards, player_2_sub_game_cards)
        elif player_1_card > player_2_card:
            winner = 'player 1'
        else:
            winner = 'player 2'
            
        if winner == 'player 1':
            player_1_cards.extend([player_1_card, player_2_card])
        else:
            player_2_cards.extend([player_2_card, player_1_card])
            
    if len(player_1_cards) == 0:
        return 'player 2'
    else:
        return 'player 1'

fill_player_decks()
winner = determine_winner(player_1_cards, player_2_cards)

if len(player_1_cards) == 0:
    print(determine_deck_score(player_2_cards))
else:
    print(determine_deck_score(player_1_cards))
```
