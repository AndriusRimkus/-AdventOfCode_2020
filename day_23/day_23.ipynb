```python
input = '952438716'
input = list(map(int, list(input)))

class Cup:
    def __init__(self, label, left_cup, right_cup):
        self.label = label
        self.left_cup = left_cup
        self.right_cup = right_cup
        
    def __repr__(self):
        if self.left_cup:
            left_cup_label = self.left_cup.label
        else:
            left_cup_label = None
        
        if self.right_cup:
            right_cup_label = self.right_cup.label
        else:
            right_cup_label = None
        
        return '{0} <- {1} -> {2}'.format(left_cup_label, self.label, right_cup_label)

    def __str__(self):
        if self.left_cup:
            left_cup_label = self.left_cup.label
        else:
            left_cup_label = None
        
        if self.right_cup:
            right_cup_label = self.right_cup.label
        else:
            right_cup_label = None
        
        return '{0} <- {1} -> {2}'.format(left_cup_label, self.label, right_cup_label)

cup_map = {}
min_cup_label = None
max_cup_label = None
active_cup = None

def read_input(input):
    global cup_map
    global min_cup_label
    global max_cup_label
    global active_cup
    
    prev_cup = None

    for cup_label in input:
        cup = Cup(cup_label, prev_cup, None)

        if prev_cup:
            prev_cup.right_cup = cup

        prev_cup = cup

        cup_map[cup_label] = cup

        if not min_cup_label or cup_label < min_cup_label:
            min_cup_label = cup_label

        if not max_cup_label or cup_label > max_cup_label:
            max_cup_label = cup_label

    cup_map[input[0]].left_cup = cup_map[input[-1]]
    cup_map[input[-1]].right_cup = cup_map[input[0]]

    active_cup = cup_map[input[0]]

def extract_three_cups(active_cup):
    extracted_cups = []
    
    current_cup = active_cup.right_cup
    
    while len(extracted_cups) < 3:
        extracted_cups.append(current_cup)
        current_cup = current_cup.right_cup
        
    active_cup.right_cup = extracted_cups[-1].right_cup
    extracted_cups[-1].right_cup.left_cup = active_cup
    
    extracted_cups[0].left_cup = None
    extracted_cups[-1].right_cup = None
    
    return extracted_cups

def include_cups(destination_cup, cups):
    destination_cup.right_cup.left_cup = cups[-1]
    cups[-1].right_cup = destination_cup.right_cup
    
    destination_cup.right_cup = cups[0]
    cups[0].left_cup = destination_cup

def determine_destination_cup(active_cup, labels_to_exclude):
    destination_label = active_cup.label - 1
    
    while True:
        if destination_label < min_cup_label:
            destination_label = max_cup_label
        
        if destination_label in labels_to_exclude:
            destination_label -= 1
            continue
        
        return cup_map[destination_label]

# part 1

read_input(input)

move = 0

while move < 100:
    extracted_cups = extract_three_cups(active_cup)

    labels_to_exclude = list(map(lambda x: x.label, extracted_cups))
    destination_cup = determine_destination_cup(active_cup, labels_to_exclude)
    
    include_cups(destination_cup, extracted_cups)

    active_cup = active_cup.right_cup
    
    move += 1
    
answer = ''
start_cup = cup_map[1]
loop_cup = start_cup.right_cup

while loop_cup != start_cup:
    answer += str(loop_cup.label)
    loop_cup = loop_cup.right_cup
    
# part 2

curr_index = max_cup_label + 1

while (curr_index <= 1000000):
    input.append(curr_index)
    curr_index += 1
    
read_input(input)

move = 0

while move < 10000000:
    extracted_cups = extract_three_cups(active_cup)

    labels_to_exclude = list(map(lambda x: x.label, extracted_cups))
    destination_cup = determine_destination_cup(active_cup, labels_to_exclude)
    
    include_cups(destination_cup, extracted_cups)

    active_cup = active_cup.right_cup
    
    move += 1
    
print(cup_map[1].right_cup.label * cup_map[1].right_cup.right_cup.label)
```
