```python
import re

input_file = open("input.txt")
input = input_file.read().strip().split('\n')
input_file.close()

idx = 0

# PART 1

def process_exp(exp):
    global idx
    
    result = 0
    last_op = '+'
    
    while (idx < len(exp)):
        if (exp[idx] in ['+', '*']):
            last_op = exp[idx]
        
        if (exp[idx] == ')'):
            return result
        
        if (exp[idx] == '(' and last_op == '+'):
            idx += 1
            result += process_exp(exp)
        
        if (exp[idx] == '(' and last_op == '*'):
            idx += 1
            result *= process_exp(exp)
        
        if (exp[idx].isnumeric() and last_op == '+'):
            result += int(exp[idx])
            
        if (exp[idx].isnumeric() and last_op == '*'):
            result *= int(exp[idx])
        
        idx += 1
    
    return result

sum = 0

for input_line in input:
    idx = 0
    exp = re.findall('\d+|[\+\*\(\)]', input_line)
    sum += process_exp(exp)

print(sum)

# PART 2

def calculate_exp_results(exp):
    while '+' in exp:
        operand_index = exp.index('+')

        result = exp[operand_index-1] + exp[operand_index+1]
        exp[operand_index] = result

        del exp[operand_index-1]
        del exp[operand_index]
    
    while '*' in exp:
        operand_index = exp.index('*')

        result = exp[operand_index-1] * exp[operand_index+1]
        exp[operand_index] = result

        del exp[operand_index-1]
        del exp[operand_index]
        
    return exp[0]

def process_exp_p2(exp):
    global idx
    local_exp = []
    
    while (idx < len(exp)):
        if (exp[idx] == ')'):            
            return calculate_exp_results(local_exp)
        
        if (exp[idx] in ['+', '*']):
            local_exp.append(exp[idx])
        
        if (exp[idx].isnumeric()):
            local_exp.append(int(exp[idx]))
        
        if (exp[idx] == '('):
            idx += 1
            
            value = process_exp_2(exp)
            local_exp.append(value)
        
        idx += 1
    
    return calculate_exp_results(local_exp)

sum = 0

for input_line in input:
    idx = 0
    exp = re.findall('\d+|[\+\*\(\)]', input_line)
    sum += process_exp_p2(exp)

print(sum)
```
