```python
import re
import math
import numpy as np
import itertools

input_file = open("input.txt")
input = input_file.read().strip().split('\n')
input_file.close()

foods = []
available_ingredients = []
global_ingredient_list = []

# PART 1

for line in input:
    data = line.strip(')').split(' (contains ')
    ingredients = data[0].split()
    allergens = data[1].split(', ')
    
    foods.append({
        'ingredients': ingredients,
        'allergens': allergens
    })
    
    global_ingredient_list.extend(ingredients)
    
    for ingredient in ingredients:
        if not ingredient in available_ingredients:
            available_ingredients.append(ingredient)

allergen_map = {}
    
for food in foods:
    for allergen in food['allergens']:
        if not allergen in allergen_map:
            allergen_map[allergen] = food['ingredients']
            continue
        
        allergen_map[allergen] = np.intersect1d(allergen_map[allergen], food['ingredients']).tolist()
        
maybe_bad_ingredients = np.concatenate(list(allergen_map.values()))
maybe_bad_ingredients = np.unique(maybe_bad_ingredients).tolist()

good_ingredients = []

answer = 0

for ingredient in available_ingredients:
    if not ingredient in maybe_bad_ingredients:
        good_ingredients.append(ingredient)
        answer += global_ingredient_list.count(ingredient)
        
# print(answer)

# PART 2

resolved_allergens = {}

def remove_ingredient_from_allergen_map(ingredient_to_remove):
    global allergen_map
    new_map = {}
    
    for allergen, ingredients in allergen_map.items():
        if ingredient_to_remove in ingredients:
            ingredients.remove(ingredient_to_remove)
        
        if len(ingredients) > 0:
            new_map[allergen] = ingredients
            
    allergen_map = new_map

while allergen_map:
    for allergen, ingredients in allergen_map.items():
        if len(ingredients) == 1:
            resolved_allergens[allergen] = ingredients[0]
            remove_ingredient_from_allergen_map(ingredients[0])
            break


sorted_allergens = {k:v for k,v in sorted(resolved_allergens.items())}
print(','.join(sorted_allergens.values()))
```
