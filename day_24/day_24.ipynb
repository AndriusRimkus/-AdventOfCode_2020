```python
import re
import math
import numpy as np
import itertools
import collections

input_file = open("input.txt")
input = input_file.read().strip().split('\n')
input_file.close()

available_directions = ['e', 'se', 'sw', 'w', 'nw', 'ne']

direction_offsets = {
    'e': (2, 0),
    'se': (1, 1),
    'sw': (-1, 1),
    'w': (-2, 0),
    'nw': (-1, -1),
    'ne': (1, -1)
}

flipped_tiles = {}

# DAY 1

for line in input:
    current_pos = [0, 0]
    
    while len(line) > 0:
        for direction in available_directions:
            if line.startswith(direction):
                line = line[len(direction):]

                offsets = direction_offsets[direction]

                current_pos[0] += offsets[0]
                current_pos[1] += offsets[1]

                if len(line) == 0:
                    key = str(current_pos[0]) + '|' + str(current_pos[1])

                    if key in flipped_tiles:
                        del flipped_tiles[key]
                    else:
                        flipped_tiles[key] = True

# print(len(flipped_tiles))    
                        
# DAY 2

# Any black tile with zero or more than 2 black tiles immediately adjacent to it is flipped to white.
# Any white tile with exactly 2 black tiles immediately adjacent to it is flipped to black.

def pass_day():
    global flipped_tiles
    flipped_tiles_updated = {}
    
    for coords in list(flipped_tiles.keys()):
        coords = list(map(int, coords.split('|')))

        count = 0

        for offset in list(direction_offsets.values()):
            key = str(coords[0] + offset[0]) + '|' + str(coords[1] + offset[1])

            if key in flipped_tiles:
                count += 1

        if count == 1 or count == 2:
            key = str(coords[0]) + '|' + str(coords[1])
            flipped_tiles_updated[key] = True

    for coords in list(flipped_tiles.keys()):
        coords = list(map(int, coords.split('|')))

        for offset in list(direction_offsets.values()):
            key = str(coords[0] + offset[0]) + '|' + str(coords[1] + offset[1])

            if key in flipped_tiles:
                continue

            white_tile_coords = (coords[0] + offset[0], coords[1] + offset[1])

            count = 0

            for offset in list(direction_offsets.values()):
                key = str(white_tile_coords[0] + offset[0]) + '|' + str(white_tile_coords[1] + offset[1])

                if key in flipped_tiles:
                    count += 1

            if count == 2:
                key = str(white_tile_coords[0]) + '|' + str(white_tile_coords[1])
                flipped_tiles_updated[key] = True
                
    flipped_tiles = flipped_tiles_updated

for i in range(0, 100):
    pass_day()

print(len(flipped_tiles))  
```
